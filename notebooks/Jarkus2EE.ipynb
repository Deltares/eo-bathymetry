```python
# !pip install git+https://github.com/openearth/jarkus.git
%matplotlib inline

import numpy as np
from jarkus.transects import Transects
import matplotlib.pyplot as plt
import pandas as pd
from netCDF4 import num2date
```

Transects in the two areas of interest are: 
- Ameland Zeegat: 
  - $\textit{Terschelling}$, id: $\texttt{4}$
  - $\textit{Ameland}$, id: $\texttt{3}$

  
- Western Scheldt delta: 
  - $\textit{Walcheren}$, id: $\texttt{16}$
  - $\textit{Zeeuws-Vlaanderen}$, id: $\texttt{17}$



```python
tr = Transects(url='http://opendap.deltares.nl/thredds/dodsC/opendap/rijkswaterstaat/jarkus/profiles/transect.nc')
ids = tr.get_data('id')
idx = np.nonzero(np.logical_or(np.logical_and(ids>=3e6,ids<5e6),
               np.logical_and(ids>=16e6,ids<18e6)))[0]
yearx = np.arange(2015,2018)
#idx
```

Running the cell below can last over a minute. It looks for the altitude and build the table.

```python
%%time
lat = []
lon = []
df_tot = pd.DataFrame()

for ix in range(len(yearx)):
        tr.set_filter(alongshore=idx, year=yearx[ix])
        z = np.squeeze(tr.get_data('altitude'))[idx,:]
        lat = tr.get_data('lat')[idx,:]
        lon = tr.get_data('lon')[idx,:]
        t = tr.get_data('time')
        # get times as datetime
        t_topo = [num2date(tt,tr.ds.variables['time_topo'].units) 
                  if ~np.isnan(tt) 
                  else np.nan 
                  for tt in tr.get_data('time_topo')[0][idx]]
        t_bathy = [num2date(tt,tr.ds.variables['time_bathy'].units) 
                  if ~np.isnan(tt) 
                  else np.nan 
                  for tt in tr.get_data('time_bathy')[0][idx]]
        print('year '+str(yearx[ix]))
        df_base = pd.DataFrame(data={'id': np.ravel(np.tile(ids[idx],(lat.shape[1],1)).T),
                              'lat': np.ravel(np.squeeze(lat)),
                              'lon': np.ravel(np.squeeze(lon))})
        
        df = pd.DataFrame(data={'z_'+str(yearx[ix]): np.ravel(z),
                              'time_topo_'+str(yearx[ix]): np.ravel(np.tile(t_topo,(lat.shape[1],1)).T),
                              'time_bathy_'+str(yearx[ix]): np.ravel(np.tile(t_bathy,(lat.shape[1],1)).T),
                             })
        df_tot = pd.concat([df_tot, df],axis=1,join='outer')
        #
# add to the base with lat, lon, id
df_tr = pd.concat([df_base, df_tot],axis=1,join='outer')
```
